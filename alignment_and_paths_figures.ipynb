<h1>Paper path and alginment generation</h1>

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore') # Hide warnings that can generate and clutter notebook

x = np.array(
    [
        -0.7553383207,
        0.4460987596,
        1.197682907,
        0.1714334808,
        0.5639929213,
        0.6891222874,
        1.793828873,
        0.06570866314,
        0.2877381702,
        1.633620422,
    ]
)

y = np.array(
    [
        0.01765193577,
        1.536784164,
        -0.1413292622,
        -0.7609346135,
        -0.1767363331,
        -2.192007072,
        -0.1933165696,
        -0.4648166839,
        -0.9444888843,
        -0.239523623,
    ]
)
```

```python
# Generate the path for each distance
from sktime.distances import distance_alignment_path
dtw_path = distance_alignment_path(x, y, metric='dtw')
dtw_path_windowed = distance_alignment_path(x, y, metric='dtw', window=0.2)
edr_path = distance_alignment_path(x, y, metric='edr')
erp_path = distance_alignment_path(x, y, metric='erp', epsilon=1.0)
lcss_path = distance_alignment_path(x, y, metric='lcss')
msm_path = distance_alignment_path(x, y, metric='msm')
twe_path = distance_alignment_path(x, y, metric='twe')
wdtw_path = distance_alignment_path(x, y, metric='wdtw')
wdtw_path_g2 = distance_alignment_path(x, y, metric='wdtw', g=0.2)
wdtw_path_g3 = distance_alignment_path(x, y, metric='wdtw', g=0.3)
```

```python
# This function does the above but generates a graph from it
from sktime._contrib._plot_path import _plot_path, _plot_alignment

plt_path = _plot_path(x, y, metric='dtw', title='dtw path window 0.2')
plt_path.show()
plt_alignment = _plot_alignment(x, y, metric='dtw', title='dtw alignment window 0.2')
plt_alignment.show()
```

```python
plt_path = _plot_path(x, y, metric='dtw', title='dtw path window 0.2', dist_kwargs={'window':0.2})
plt_path.show()
plt_alignment = _plot_alignment(x, y, metric='dtw', title='dtw alignment window 0.2', dist_kwargs={"window": 0.2})
plt_alignment.show()
```
